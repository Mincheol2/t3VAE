```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('sp500_historical.csv')
ratio = []
for i in range(len(data)-1):
    ratio.append((data['Close'][i+1] - data['Close'][i]) / data['Close'][i])
```

```python
plt.plot(ratio)
plt.title("S&P500 Daily Changes")
plt.ylabel("Change (%)")
plt.show()
```

```python
_ = sns.displot(ratio,kind="kde")
```
